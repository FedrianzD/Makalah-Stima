```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
df0 = pandas.read_csv("../test/test.csv")
df = df0[:6]
df2 = df0[6:]
df2
```

```python
sns.lineplot(x = df['Panjang Text'], y = df['KMP'], color=sns.color_palette("muted")[0], label='KMP')
sns.lineplot(x = df['Panjang Text'], y = df['BM'], color=sns.color_palette("muted")[1], label='BM')
sns.lineplot(x = df['Panjang Text'], y = df['Aho-Corasick'], color=sns.color_palette("muted")[2], label='Aho-Corasick')
plt.title("Panjang Text Terhadap Waktu Eksekusi")
plt.ylabel("Waktu Eksekusi (\u00B5s)")
plt.legend()
```

```python
sns.lineplot(x = df2['Jumlah Pattern'], y = df2['KMP'], color=sns.color_palette("muted")[0], label='KMP')
sns.lineplot(x = df2['Jumlah Pattern'], y = df2['BM'], color=sns.color_palette("muted")[1], label='BM')
sns.lineplot(x = df2['Jumlah Pattern'], y = df2['Aho-Corasick'], color=sns.color_palette("muted")[2], label='Aho-Corasick')
plt.title("Jumlah Pattern Terhadap Waktu Eksekusi")
plt.ylabel("Waktu Eksekusi (\u00B5s)")
plt.legend()
```

```python
df3 = pandas.read_csv("../test/test2.csv")
df3
```

```python
sns.lineplot(x = df3['no'], y = df3['KMP'], color=sns.color_palette("muted")[0], label='KMP')
sns.lineplot(x = df3['no'], y = df3['BM'], color=sns.color_palette("muted")[1], label='BM')
sns.lineplot(x = df3['no'], y = df3['Aho'], color=sns.color_palette("muted")[2], label='Aho-Corasick')
plt.title("Panjang Text dan Jumlah Pattern Terhadap Waktu Eksekusi")
plt.ylabel("Waktu Eksekusi (\u00B5s)")
plt.xlabel("(Panjang Text, Jumlah Pattern)")

custom_ticks = ['(1000,1)', '(2000,2)', '(4000,4)', '(8000,8)', '(16000,16)', '(32000,32)']
plt.xticks(df3['no'], custom_ticks)
plt.legend()
```
